```python
import pandas as pd
data_frame = pd.read_csv('233_2.3_iris.csv')
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LinearRegression
```

```python
df = data_frame[['Petal.Length', 'Petal.Width']]
```

```python
train_set, test_set = train_test_split(df, test_size=0.2, random_state=42)
```

```python
df_copy = train_set.copy()
```

```python
df_copy.corr()
```

R mesura com estan de relacionades entre si dues variables. Com més proper a 1 o a -1, més estreta será la relació. 

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df_copy['Petal.Length'], df_copy['Petal.Width'])
```

```python
train_set_x = df_copy.drop(['Petal.Width'], axis = 1)
train_set_label = df_copy['Petal.Width']
```

```python
lin_reg = LinearRegression()
lin_reg.fit(train_set_x, train_set_label)
```

```python
print('Coefficients:', lin_reg.coef_)
print('Intercept: ', lin_reg.intercept_)
```

Petal.width = 0.41323829*Petal Length - 0.35666804105655303

```python
lin_reg.predict(4)
```

```python
# Assuming you're using numpy, reshape your input to be 2D array
import numpy as np

# Correct way to reshape the input for a single feature
value_to_predict = np.array([[4]])

# Now use the reshaped array for prediction
predicted_value = lin_reg.predict(value_to_predict)

print(predicted_value)
```

Valor de l'amplada que hi prediu = 1.296

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

# Assuming x and y are already defined
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0)

scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)

n_neighbors = 7
knnModel = KNeighborsClassifier(n_neighbors)
knnModel.fit(x_train, y_train)  # Use x_train and y_train instead of X_train and t_train
```

```python
# Correcting the variable names for fitting the model
knnModel.fit(x_train, y_train)  # Use x_train and y_train

# Correcting the print statement to use the proper format specifier
print('Accuracy del model KNN pel train set és: {:.2f}'.format(knnModel.score(x_train, y_train)))
```

La precisió pel paquet d'entrenament o train set és del 90%.
Fem servir un model de classificació perquè volem predir una categoria (1,2,3,4 o 5) i no un número de valor quantitatiu.

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
pred = knnModel.predict(x_test)
print(confusion_matrix(y_test, pred))
print(classification_report(y_test,pred))
```

0.87 és prou bona
