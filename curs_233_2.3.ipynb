- El coeficient  de  determinació, conegutcom a 푅²,  indica  quina  quantitat  de  variació  en yes  pot explicar  per a la  dependència  d’퐱mitjançant  el  model  de  regressió.El  major R2 indica  un millor  ajustament  i  significa  que  el  model  pot  explicar  millor  la  variació  de  la  sortida  amb diferents  entrades. - underfitting: coeficient  de  correlació massa  baix
- overfitting: coeficient  de  correlació massa  alts. Això voldrà dir que el  model  encaixaperfectament  amb  les  dades  donades,  però  corre  el  risc  de  preveure  molt malament    quan    s’aplica    a    altres    dades    amb    les    quals    no    ha    estat    entrenat
